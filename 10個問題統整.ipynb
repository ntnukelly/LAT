7. 台大、師大、台科大所學校各年的外國學生總數差異?

```python
data_ntnu = data[data['school'] == '國立臺灣師範大學']
data_ntu = data[data['school'] == '國立臺灣大學']
data_ntust = data[data['school'] == '國立臺灣科技大學']

# 以年份分組，計算各學校外國學生總數
data_ntnu = data_ntnu.groupby('year')['subtotal'].sum()
data_ntu = data_ntu.groupby('year')['subtotal'].sum()
data_ntust = data_ntust.groupby('year')['subtotal'].sum()

# 合併三所學校的資料
data_combined = pd.concat([data_ntnu, data_ntu, data_ntust], axis=1, keys=['國立臺灣師範大學', '國立臺灣大學', '國立臺灣科技大學'])

# 顯示結果
print(data_combined)
```

8. 台大、師大、台科大所學校各年最多外國學生的科系為何?

```python
ntu = data[data['school'] == '國立臺灣大學']
ntnu = data[data['school'] == '國立臺灣師範大學']
ntust = data[data['school'] == '國立臺灣科技大學']

# 按照學年和系所分組，取得每年最多外國學生的系所
ntu_top_dept = ntu.groupby(['year', 'department name'])['subtotal'].sum().groupby(level=0).idxmax()
ntnu_top_dept = ntnu.groupby(['year', 'department name'])['subtotal'].sum().groupby(level=0).idxmax()
ntust_top_dept = ntust.groupby(['year', 'department name'])['subtotal'].sum().groupby(level=0).idxmax()

# 將每年最多外國學生的系所結果合併
top_dept = pd.concat([ntu_top_dept, ntnu_top_dept, ntust_top_dept], axis=1)
top_dept.columns = ['國立臺灣大學', '國立臺灣師範大學', '國立臺灣科技大學']

# 顯示結果
print(top_dept)
```

9. 師大華語系各年的外國學生在學比例為何?

```python
# 篩選出師大華語系的資料
data_shida = data[(data['school'] == '國立臺灣師範大學') & (data['department name'] == '華語文教學系')]

# 計算各年華語系外國學生在學比例
data_ratio = data_shida[['year', 'enrollment ratio']].groupby('year').mean()
```
